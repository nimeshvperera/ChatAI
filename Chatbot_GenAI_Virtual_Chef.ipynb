```python
import os
from google import genai
from google.genai import types
```

```python
client = genai.Client(api_key=IP_COACHING)
```

```python
trainer_prompt = """
You are Chef Buddy, a warm, fun, and encouraging cooking assistant. You only discuss food, cooking, and recipes. If asked anything else, redirect back to cooking.

Before suggesting any recipe, always ask:
1. Any dietary restrictions or preferences?
2. Cooking skill level — beginner, intermediate, or advanced?

Never suggest a recipe until both are answered.

When giving a recipe:
- Present ingredients as a numbered list with related emojis for each ingredient where possible
- Present steps as a numbered list with clear instructions
- Beginners get a maximum of 6 ingredients
- Always explain any culinary jargon in simple terms

Never give exact calorie counts — if asked, say: "I'm a Chef Buddy (obese), not a nutritionist!"

Keep every response warm, encouraging, and fun.


"""



```

```python
try:
    chat = client.chats.create(
        model="gemini-2.5-flash",
        config=types.GenerateContentConfig(
            system_instruction=trainer_prompt,
            temperature=0.7,
        )
    )

    print("Your Friendly Cooking Assistant is ready! (Type 'quit' or 'exit' to stop)", flush=True)
    print("-" * 65, flush=True)

    while True:
        user_input = input("You: ")

        if user_input.lower() in ['quit', 'exit']:
            print("\nChef: Great session today! Keep up the good work, stay cookin', good lookin'!", flush=True)
            break

        if not user_input.strip():
            continue

        try:
            response = chat.send_message(user_input)

            print(f"\n{'='*65}", flush=True)
            print(f"User's Question:\n  {user_input}", flush=True)
            print(f"\nBot's Answer:\n  {response.text.strip()}", flush=True)
            print(f"{'='*65}\n", flush=True)

        except Exception as e:
            print(f"\n[Error] The API returned an error: {e}\n", flush=True)

except Exception as e:
    print(f"Failed to initialize client. Check your API key.\nError: {e}", flush=True)
```
