```python
import os
from google import genai
from google.genai import types
```

```python
IP_COACHING = "AIzaSyAyuNV3SwNA5-ZuRBiJxmdVJGAj_Vclq58"
```

```python
client = genai.Client(api_key=IP_COACHING)
```

```python
trainer_prompt = """
The goal is to help students learn any academic topic.
I did this by:
- Using only encouraging and gentle tones
- Answer one question at a time and confirm understanding before proceeding to next topic/question
- Quizzing students after explanation to make sure they understand
- Encourage if an incorrect answer is given and provide hints before providing the correct one
- Direct questions to homework are NOT allowed
- Start with simple explanations first before going into detail
- If technical terms/jargon are used, explain in simple terms

"""



```

```python
try:
    chat = client.chats.create(
        model="gemini-2.5-flash",
        config=types.GenerateContentConfig(
            system_instruction=trainer_prompt,
            temperature=0.7,
        )
    )

    print("Your Friendly Study Buddy is ready! (Type 'quit' or 'exit' to stop)", flush=True)
    print("-" * 65, flush=True)

    while True:
        user_input = input("You: ")

        if user_input.lower() in ['quit', 'exit']:
            print("\nCoach: Great session today! Keep up the good work\'!", flush=True)
            break

        if not user_input.strip():
            continue

        try:
            response = chat.send_message(user_input)

            print(f"\n{'='*65}", flush=True)
            print(f"User's Question:\n  {user_input}", flush=True)
            print(f"\nBot's Answer:\n  {response.text.strip()}", flush=True)
            print(f"{'='*65}\n", flush=True)

        except Exception as e:
            print(f"\n[Error] The API returned an error: {e}\n", flush=True)

except Exception as e:
    print(f"Failed to initialize client. Check your API key.\nError: {e}", flush=True)
```
